```python
# TASK 13 - What is the most commonly used title? (find this from the Name attribute)

df_titanic["Title"] = df_titanic["Name"].apply(lambda x: x.split(',')[1].split(" ")[1]) 
# split name to get titles
df_titanic.head(5)

df_titanic['Title'].mode()

df_titanic['Title'].nunique()

# My method
import re

# Extract titles using regex pattern
pattern = r',\s*([A-Za-z]+)\..*'  # Word between comma and period
titles = df_titanic['Name'].str.extract(pattern, expand=False)
title_counts = titles.value_counts()

title_me("Title distribution")
for title, count in title_counts.items():
    print( f"{title}: {count}" )
    
print( f"\nMost common title: '{title_counts.index[0]}'" )
# -> 'Mr'
```
